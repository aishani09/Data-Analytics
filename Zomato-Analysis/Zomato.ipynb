```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Check unique values in the 'Rating color' column
unique_rating_colors = df['Rating color'].unique()
print("Unique Rating Colors:", unique_rating_colors)

# Define a mapping of rating colors to their respective colors
color_mapping = {
    'Red': 'red',               # Poor
    'Orange': 'orange',         # Average
    'Yellow': 'yellow',         # Good
    'Green': 'green',           # Very Good
    'Dark Green': 'darkgreen',  # Excellent
    'Grey': 'grey',             # No rating
    'White': 'lightgrey'        # Example additional mapping for 'White'
    # Add any other colors that may be present in the dataset
}

# Count the number of restaurants per rating color
rating_color_counts = df['Rating color'].value_counts()

# Create a color list based on the index of the rating colors
colors = [color_mapping[color] for color in rating_color_counts.index]

# Plot the distribution
plt.figure(figsize=(10,6))
sns.barplot(x=rating_color_counts.index, y=rating_color_counts.values, palette=colors)
plt.title('Distribution of Restaurants Based on Rating Color')
plt.xlabel('Rating Color')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
from collections import Counter

cuisine_list = df['Cuisines'].dropna().str.split(', ').sum()
top_cuisines = Counter(cuisine_list).most_common(10)
cuisines_df = pd.DataFrame(top_cuisines, columns=['Cuisine', 'Count'])
plt.figure(figsize=(10,6))
sns.barplot(x='Cuisine', y='Count', data=cuisines_df)
plt.title('Top 10 Cuisines Offered')
plt.xticks(rotation=45)
plt.show()
```

```python
selected_city = input("Enter the city name to filter the top 10 restaurants: ")

# Convert the rating column to numeric (if necessary)
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Drop rows with NaN values in 'Rating' or 'City' if any
df = df.dropna(subset=['Aggregate rating', 'City'])

# Filter the dataframe for the selected city
filtered_df = df[df['City'].str.lower() == selected_city.lower()]  # Case-insensitive match

# Check if there are any restaurants in the selected city
if filtered_df.empty:
    print(f"No restaurants found in {selected_city}. Please check the city name.")
else:
    # Sort the filtered dataframe by 'Rating' in descending order and get the top 10 restaurants
    top_10_restaurants = filtered_df.sort_values(by='Aggregate rating', ascending=False).head(10)

    # Display the top 10 restaurants
    print(f"Top 10 Restaurants in {selected_city}:")
    print(top_10_restaurants[['Restaurant Name', 'Aggregate rating']])
```
